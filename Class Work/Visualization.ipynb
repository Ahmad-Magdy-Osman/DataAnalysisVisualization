## Finding the Adjusted Gross by Year Per Production Company (the top 4 only)

To do this efficiently we will join two dataframes and then we will need to group by and sum to get the total In the end I would like to display a stacked area chart.

```python
top4 = movieProd.groupby(["Production"]).sum().sort_values("AdjustedGross", ascending=False).head(4)
```

```python
top4
```

```python
top4.index
```

```python
top4ProdMovies = movies.loc[movies["Production"].isin(top4.index)]
```

```python
top4ProdMovies
```

```python
top4yearAdjustGross = top4ProdMovies[["Production", "Year", "AdjustedGross"]]
```

```python
top4yearAdjustGross
```

```python
top4yearAdjustGross.set_index("Production", inplace=True)
```

```python
top4yearAdjustGross
```

```python
top4ByYear = top4yearAdjustGross.groupby(["Production", "Year"]).sum()
```

```python
top4ByYear
```

```python
top4ByYear.reset_index(level=0, inplace=True)
```
