```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
vehicles = pd.read_csv("cars.csv")
```

```python
# How many of those?
vehicles.state_name == "FAILED"
```

```python
# Removing canadian cars
vehicles = vehicles[vehicles.state_name != "FAILED"]
```

```python
vehicles.shape
```

We have about 58833 cars from Canada... Which we just removed.

```python
vehicles.price = vehicles.price[~((vehicles.price-vehicles.price.mean()).abs() > .05*vehicles.price.std())]
vehicles.odometer = vehicles.odometer[~((vehicles.odometer-vehicles.odometer.mean()).abs() > 5*vehicles.odometer.std())]
vehicles.year = vehicles.year[~((vehicles.year-vehicles.year.mean()).abs() > 10*vehicles.year.std())]
```

We just normalized values for ourselves. LOVELY!

* The white dot represents the median
* The thick gray bar in the center represents the interquartile range
* The thin gray line represents the 95% confidence interval
* On each side of the gray line is a kernel density estimation to show the distribution shape of the data. 
* Wider sections of the violin plot represent a higher probability that members of the population will take on the given value;
* The skinnier sections represent a lower probability.

```python
x = "odometer"
y = "price"

xQuantiles = [0]
data = {}
for i in range(10):
    xQuantiles.append(vehicles[x].quantile((i+1)/10))
for i in range(10):
    xQuantilesFrame = vehicles[vehicles.odometer.between(xQuantiles[i], xQuantiles[i+1])]
    yQuantiles = [0]
    yMeans = []
    for j in range(10):
        yQuantiles.append(xQuantilesFrame[y].quantile((j+1)/10))
    for j in range(10):
        yQuantilesFrame = xQuantilesFrame[xQuantilesFrame[y].between(yQuantiles[j], yQuantiles[j+1])]
        yMean = yQuantilesFrame[y].mean()
        yMeans.append(round(yMean))
    data["{}-{}".format(round(xQuantiles[i]), round(xQuantiles[i+1]))] = yMeans
percentileFrame = pd.DataFrame(data)
print(percentileFrame)
```

```python
conditions = ["new", "like new", "excellent", "good", "fair", "salvage"]
fig, ax = plt.subplots(figsize=(20, 20), nrows = 6, ncols = 2)
loc = -1
rounded = vehicles.copy()
rounded.price = vehicles.price.round(decimals=-3)
rounded.odometer = vehicles.odometer.round(decimals=-3)
for i in conditions:
    loc += 1
    catData = rounded[rounded["condition"].values == i]
    median = catData.groupby("odometer")["price"].median()
    mean = catData.groupby("odometer")["price"].mean()
    medianRolling = median.rolling(15).median()
    meanRolling = mean.rolling(15).mean()
    xMedian = medianRolling.index.values
    yMedian = medianRolling.values
    xMean = meanRolling.index.values
    yMean = meanRolling.values
    ax[loc][0].set_xlim([0,200000])
    ax[loc][0].set_ylim([0,30000])
    ax[loc][1].set_xlim([0,200000])
    ax[loc][1].set_ylim([0,30000])
    ax[loc][0].plot(xMedian, yMedian)
    ax[loc][1].plot(xMean, yMean)
    ax[loc][0].set_title("Means: {}".format(i))
    ax[loc][1].set_title("Medians: {}".format(i))
```
